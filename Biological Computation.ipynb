```python
import os
import glob
import pandas as pd



data_folder = r"C:\Users\athar\Downloads\49 mutants"


file_pattern = os.path.join(data_folder, "*_data_file*.txt")


all_data = []


for file_path in glob.glob(file_pattern):
    try:
        df = pd.read_csv(file_path, sep="\t") 
        df['SourceFile'] = os.path.basename(file_path)
        all_data.append(df)
    except Exception as e:
        print(f"Error reading {file_path}: {e}")


if all_data:
    combined_df = pd.concat(all_data, ignore_index=True)
    print(" Successfully loaded all gene files!")
    display(combined_df)  
else:
    print(" No files found. Check your path or file names.")
```

```python


def compute_avg(row, prefix):
    return (row[f"{prefix}.Rep1"] + row[f"{prefix}.Rep2"] + row[f"{prefix}.Rep3"]) / 3


combined_df["Avg_mRNA_WT"] = combined_df.apply(lambda row: compute_avg(row, "mRNA.Expression.WT"), axis=1)
combined_df["Avg_mRNA_Mut"] = combined_df.apply(lambda row: compute_avg(row, "mRNA.Expression.Mut"), axis=1)

combined_df["Avg_Protein_WT"] = combined_df.apply(lambda row: compute_avg(row, "Protein.Expression.WT"), axis=1)
combined_df["Avg_Protein_Mut"] = combined_df.apply(lambda row: compute_avg(row, "Protein.Expression.Mut"), axis=1)

combined_df["Avg_Viability_WT"] = combined_df.apply(lambda row: compute_avg(row, "CellViability.WT"), axis=1)
combined_df["Avg_Viability_Mut"] = combined_df.apply(lambda row: compute_avg(row, "CellViability.Mut"), axis=1)


combined_df["Delta_mRNA"] = combined_df["Avg_mRNA_Mut"] - combined_df["Avg_mRNA_WT"]
combined_df["Delta_Protein"] = combined_df["Avg_Protein_Mut"] - combined_df["Avg_Protein_WT"]
combined_df["Delta_Viability"] = combined_df["Avg_Viability_Mut"] - combined_df["Avg_Viability_WT"]


combined_df[[
    "Gene", 
    "Avg_mRNA_WT", "Avg_mRNA_Mut", "Delta_mRNA",
    "Avg_Protein_WT", "Avg_Protein_Mut", "Delta_Protein",
    "Avg_Viability_WT", "Avg_Viability_Mut", "Delta_Viability"
]]
```

```python
from difflib import SequenceMatcher


def classify_mutation(wt_seq, mut_seq):
    if wt_seq == mut_seq:
        return "No Change"
    if len(wt_seq) == len(mut_seq):
        return "Substitution"
    elif len(wt_seq) < len(mut_seq):
        return "Insertion"
    else:
        return "Deletion"


def classify_location(wt_seq, mut_seq, promoter_length=1000):
    return "Promoter" if wt_seq[:promoter_length] != mut_seq[:promoter_length] else "Coding"


combined_df["MutationType"] = combined_df.apply(
    lambda row: classify_mutation(row["WildType.Sequence"], row["Mutant.Sequence"]), axis=1)

combined_df["MutationRegion"] = combined_df.apply(
    lambda row: classify_location(row["WildType.Sequence"], row["Mutant.Sequence"]), axis=1)


combined_df[["Gene", "MutationType", "MutationRegion"]]
```

```python

combined_df["Rank_mRNA"] = combined_df["Delta_mRNA"].rank(ascending=False)
combined_df["Rank_Protein"] = combined_df["Delta_Protein"].rank(ascending=False)
combined_df["Rank_Viability"] = combined_df["Delta_Viability"].rank(ascending=False)


combined_df["Benefit_Score"] = (
    combined_df["Rank_mRNA"] + 
    combined_df["Rank_Protein"] + 
    combined_df["Rank_Viability"]
)


top_5 = combined_df.sort_values("Benefit_Score").head(5)


top_5[[
    "Gene", "MutationType", "MutationRegion",
    "Delta_mRNA", "Delta_Protein", "Delta_Viability",
    "Benefit_Score"
]]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


sns.set(style="whitegrid")


top_genes = ["Avon4", "Avon1a5", "Jams9r2", "Cairn3a2", "Gorm6b6"]
highlight_df = combined_df[combined_df["Gene"].isin(top_genes)]

# === 1. SCATTER PLOT: Protein vs Viability ===
plt.figure(figsize=(8, 6))
sns.scatterplot(
    data=combined_df,
    x="Delta_Protein",
    y="Delta_Viability",
    hue="MutationType",
    s=100,
    edgecolor="w",
    alpha=0.6
)


sns.scatterplot(
    data=highlight_df,
    x="Delta_Protein",
    y="Delta_Viability",
    color="red",
    marker="*",
    s=250,
    label="Top 5 Genes",
    edgecolor="black"
)


for _, row in highlight_df.iterrows():
    plt.text(row["Delta_Protein"] + 0.5, row["Delta_Viability"], row["Gene"], fontsize=9)

plt.title(" Protein vs  Viability (Top 5 Genes Highlighted)")
plt.xlabel(" Protein Expression")
plt.ylabel(" Cell Viability")
plt.legend(title="Mutation Type")
plt.tight_layout()
plt.savefig("scatter_plot_highlighted_top5.png", dpi=300)
plt.show()

# === 2. BOXPLOT: Viability by Mutation Region ===
plt.figure(figsize=(8, 6))
sns.boxplot(
    data=combined_df,
    x="MutationRegion",
    y="Delta_Viability",
    palette="Set2"
)


sns.stripplot(
    data=highlight_df,
    x="MutationRegion",
    y="Delta_Viability",
    color="red",
    marker="*",
    size=15,
    jitter=True,
    label="Top 5 Genes"
)

plt.title(" Viability by Mutation Region (Top 5 Highlighted)")
plt.xlabel("Mutation Region")
plt.ylabel(" Cell Viability")
plt.legend()
plt.tight_layout()
plt.savefig("viability_boxplot_highlighted_top5.png", dpi=300)
plt.show()

# === 3. HEATMAP: Delta values with top 5 visually grouped ===

heatmap_data = combined_df.set_index("Gene")[["Delta_mRNA", "Delta_Protein", "Delta_Viability"]]

heatmap_data = heatmap_data.loc[combined_df.sort_values("Benefit_Score")["Gene"]]


plt.figure(figsize=(10, 12))
sns.heatmap(heatmap_data, cmap="coolwarm", linewidths=0.5, linecolor="gray", cbar=True)

plt.title("Heatmap of  mRNA, Protein, and Viability (Top 5 Sorted at Top)")
plt.tight_layout()
plt.savefig("heatmap_highlighted_top5.png", dpi=300)
plt.show()
```
